```python
dataset=df.drop(['food','servings'],axis=1)
dataset
```

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model=RandomForestClassifier(random_state=10)
random_forest_model.fit(X_train,Y_train.ravel())
```

```python
predict_train_data=random_forest_model.predict(X_test)
```

```python
predict_train_data
```

```python
from sklearn.metrics import r2_score
score=r2_score(Y_test,predict_train_data)
print(score)
```

```python
age=12
bloodglucose=90
Cho=95
TDDI=24
ISF=75

scaler.transform([[age,bloodglucose,Cho,TDDI,ISF]])
```

```python
value=random_forest_model.predict(scaler.transform([[age,bloodglucose,Cho,TDDI,ISF]]))
value
```
